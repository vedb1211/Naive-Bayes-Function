```python
# Naive Bayes Classifier function - devdas
# B-Tech DS - Sem III - Managing Uncertainity (MU)
# J009 Ved Bhatt & J016 Aditya Das

import numpy as np
import pandas as pd
import math as m
import matplotlib.pyplot as plt

def devdas(df,y,inp):
    
    a=df[y].to_list()
    dec=':'
    factor=[]
    if dec in inp:
        x=inp.split(':')
        for i in range(int(x[0]),int(x[1])+1):
            factor.append(fac[i])
    else:
        x=inp.split(',')
        for i in x:
            factor.append(fac[int(i)])
    factor.append(y)
    df=df[factor]
    df = df.sort_index(axis=1)
    cols=list(df.columns)
    
    mean=df.pivot_table(index=y,values=cols,aggfunc=np.mean)
    std=df.pivot_table(index=y,values=cols,aggfunc=np.std)
    std=std.to_numpy()
    mean=mean.to_numpy()
    p=m.pi
    e=m.e
    def normdist(x,m1,s1):
        k=1/((pow(2*p,0.5))*s1)
        prob=k*pow(e,-0.5*pow((x-m1)/(s1),2))
        return prob
    
    f=dict()
    for i in range(len(a)):
        if a[i] in f: f[a[i]]+=1
        else:f[a[i]]=1
    keys=list(f.keys())
    val=list(f.values())
    val=np.array(val)
    val=val/len(df)
    f=dict(zip(keys,val))
    p1=list(f.values())
    f=dict(sorted(f.items()))
    keys=list(f.keys())
    df.drop(y,axis=1,inplace=True)
    cols=list(df.columns)
    for w in keys:
        pro=[]
        for i in range(len(df)):
            l=df.iloc[i].to_list()
            temp=1
            for j in range(len(cols)):
                temp*=normdist(l[j],mean[keys.index(w),j],std[keys.index(w),j])
            pro.append(temp)
        pro=np.array(pro)
        pro=list(pro*f[w])
        df[w]=pro

    m1=[]
    t=[]
    for i in range(len(df)):
        l=df.iloc[i].to_list()[len(cols):len(df.columns)]
        test=max(l)
        t.append(test)
        index=l.index(test)
        m1.append(keys[index])
    count=0
    for i in range(len(df)):
        if m1[i]==a[i] : count+=1
    return count/len(df)

x = input("Enter your file name: ")
df=pd.read_csv(x)
fac=list(df.columns)
print('Features in dataset are:')
print(fac)
y = input("Enter your target name: ")
print('Features names:')
for i in range(len(fac)):
    print(i,' - ',fac[i])
inp=input('Input your variable features: ')
train=df.sample(frac=0.80)
test=df.drop(train.index)
trainp = devdas(train,y,inp)
testp= devdas(test,y,inp)
print("Naive Bayes accuracy for training dataset:",trainp*100,"%")
print("Naive Bayes accuracy for testing dataset:",testp*100,"%")
```
